```python
# Import API call packages
import pandas as pd
import requests
import json 
```

```python
# Input API url to send requests
url = "https://api.covid19api.com/total/country/singapore/status/confirmed"

# API request
response = requests.request("GET", url)
parsed_data = json.loads(response.text)
df = pd.DataFrame.from_dict(parsed_data)

# Cases = total cases, adding extra column for daily cases
df['Daily_cases'] = df['Cases'].diff()
df.Cases.fillna(0).sum()
df.Daily_cases.fillna(0).sum()
df.sample(20)
```

```python
# Clean up the datetime format and break it down for easier dashboarding
df['date'] = pd.DatetimeIndex(df['Date']).strftime('%y-%m-%d')
df['day'] = pd.DatetimeIndex(df['Date']).day
df['month'] = pd.DatetimeIndex(df['Date']).month
df['year'] = pd.DatetimeIndex(df['Date']).year
df['quarter'] = pd.DatetimeIndex(df['Date']).quarter
df['year/month'] = pd.DatetimeIndex(df['Date']).strftime('%y-%m')
```

```python
# Drop all unwanted columns
df_clean = df.drop(['Date', 'CountryCode', 'Province', 'City', 'CityCode', 'Lat', 'Lon'], axis = 1)

# Quick scan on the final df before exporting to csv
df_clean.sample(20)
```
